```python
df_el = load_dataset('elastic_tensor_2015', data_home = '.')
df_el
```

```python
# 可视化
import matplotlib.pyplot as plt

plt.hist(df_el['K_VRH'], color = 'g', edgecolor = 'w', alpha = 0.75, bins = [0, 50, 100, 150, 200, 250, 300, 350, 400]) #bins为统计区间
plt.show()
```

```python
# 归一化，总面积为1
import matplotlib.pyplot as plt


plt.figure(figsize=(8,8))
plt.hist(df_el['K_VRH'], rwidth=0.8, linewidth=5, density = True, color = 'g', edgecolor = 'y', alpha = 0.75, bins = [0, 50, 100, 150, 200, 250, 300, 350, 400]) #bins为统计区间
plt.show()
```

```python
import numpy as np

x = np.array(df_el['G_VRH'])
y = np.array(df_el['K_VRH'])
z = np.array(df_el['poisson_ratio'])
plt.xlim(0, 300)
plt.scatter(x, y, c = z, edgecolor = 'k', cmap = 'bwr')
plt.colorbar()
plt.show()
```

```python
import plotly.express as px

fig = px.scatter(df_el, x = 'G_VRH', y = 'K_VRH', color = 'poisson_ratio', color_continuous_scale = 'emrld', range_x = [0, 300])
fig.show()
```

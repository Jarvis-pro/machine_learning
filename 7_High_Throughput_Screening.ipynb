```python
from matminer.datasets import load_dataset

df_mp = load_dataset('mp_nostruct_20181018', data_home = '.')
df_mp
```

# 去掉缺失数据

```python
num = df_mp.isna().sum() #查找缺失数据
num
```

```python
df_mp[df_mp['formula'].isna()] 
```

```python
df_mp = df_mp.dropna(axis = 0, subset = ['formula']) #去掉缺失数据
df_mp
```

# 筛选含锂材料

```python
df_mp_Li = df_mp.loc[df_mp['formula'].str.contains('Li')] #将formula转换为字符串，筛选含Li的材料
df_mp_Li
```

```python
from matminer.featurizers.conversions import StrToComposition #将formula按照元素划分

df_test = StrToComposition().featurize_dataframe(df_mp_Li, 'formula')
df_test
```

```python
for i, row in df_test.iterrows():
    if row['composition'].get_atomic_fraction('N') == 0:
        df_test = df_test.drop([i])

df_test
```

# 筛选带隙大于0.5的材料

```python
df_mp_Li_gp = df_mp_Li.loc[df_mp_Li['gap pbe']>0.5]
df_mp_Li_gp
```

# 筛选hull小于0.005

```python
df_mp_Li_gp_hull = df_mp_Li_gp.loc[df_mp_Li_gp['e_hull']<0.005]
df_mp_Li_gp_hull
```

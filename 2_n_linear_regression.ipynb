```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('mp.csv')
df
```

```python
X = df.drop(['formula','y'], axis = 1)
y = df['y'].values
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X,y)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
print(lr.score(X, y))
```

```python
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse

print('MAE = %.3f' % mae(y_true = y, y_pred = lr.predict(X)))
print('MSE = %.3f' % mse(y_true = y, y_pred = lr.predict(X)))
```

```python
X.shape
```

```python
#留出法
X_train, y_train = X[:-10], y[:-10] #切片，从头取到倒数第十一个
X_test, y_test = X[-10:], y[-10:] #从倒数第十个取到最后一个
```

```python
X_train.shape, X_test.shape
```

```python
lr_lo = LinearRegression()
lr_lo.fit(X_train, y_train)
```

```python
R2_lo = lr_lo.score(X_test, y_test)
print('R2 = ', R2_lo)
print('RMSE = %.3f' % np.sqrt(mse(y_true = y_test, y_pred = lr_lo.predict(X_test))))
```

```python
#多次留出法，使用shuffle
```

```python
#交叉验证
from sklearn.model_selection import cross_val_score

rmse_scores = cross_val_score(lr, X, y, scoring = 'neg_root_mean_squared_error', cv = 10) #10折交叉验证

print('Folds: %i, mean RMSE:%.3f' % (len(rmse_scores), -np.mean(rmse_scores)))
```

```python
rmse_scores
```

```python
#留一法，交叉法中cv = len(X)，或者手打。效果好，但是训练量大
from sklearn.model_selection import LeaveOneOut

sample_x = [1, 2, 3, 4, 5]
loo = LeaveOneOut()
for train, test, in loo.split(sample_x):
    print(train, test)
```

```python
#自助法。取样测试放回，再取没有被抓到的样测试

from sklearn.utils import resample

X_train, y_train = resample(X, y)

index_test = set(X.index) - set(X_train.index)
```

```python
X_train.shape, y_train.shape
```

```python
X_test = X.iloc[list(index_test)]
y_test = y[list(index_test)]
```

```python
X_test.shape, y_test.shape
```

```python
lr_bs = LinearRegression()
lr_bs.fit(X_train, y_train)
```

```python
print('MSE = %.3f'% np.sqrt(mse(y_true=y_test, y_pred=lr_bs.predict(X_test))))
```

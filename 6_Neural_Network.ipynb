```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('HER2.csv')
df
```

```python
X = df.drop('Eads(H)', axis = 1)
y = df['Eads(H)']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
X_train_s = (X_train - X_train.mean()) / (X_train.max() - X_train.min())
X_test_s = (X_test - X_train.mean()) / (X_train.max() - X_train.min())
```

```python
nn = MLPRegressor(hidden_layer_sizes = (64, 64), random_state = 1)
```

```python
nn.fit(X_train_s, y_train)
```

```python
plt.figure(figsize = (5, 5))
plt.scatter(y_train, nn.predict(X_train_s), c = 'cyan', edgecolor = 'k')
plt.scatter(y_test, nn.predict(X_test_s), c = 'pink', edgecolor = 'k')
plt.plot([-1.0, 2.0], [-1.0, 2.0], 'k--')
plt.xlim(-1, 2)
plt.ylim(-1, 2)
plt.show()
```
